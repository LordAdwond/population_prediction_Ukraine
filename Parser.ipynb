```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
def get_html(url):
    return requests.get(url).text
def get_data(url):
    return BeautifulSoup(get_html(url), 'lxml')
```

Population data

```python
population_data = get_data("https://index.minfin.com.ua/reference/people/")
```

```python
table = population_data.find_all("table")[1]
```

```python
print(type(table))
```

```python
print(table)
```

```python
population = [float(value.text.replace(",", ".")[:2]+value.text.replace(",", ".")[3:]) for value in table.find_all("big")]
population
```

```python
population_data = pd.DataFrame()
population_data["year"] = [1990+i for i in range(len(population))]
population_data["population (thousands)"] = population
population_data.head()
```

```python
population_data.to_csv("population_info_primary.csv")
```

Mean salary by year in gryvnias.

```python
salary_data = get_data("https://uk.wikipedia.org/wiki/Середня_заробітна_плата_в_Україні").find_all("table")
salary_data = salary_data[-1]
```

```python
tds = salary_data.find_all("td")
print( tds )
```

```python
mid_salary = [ tds[i].text.replace(",", ".") for i in range(len(tds)-1) if i%5==1 ][4:]
print(mid_salary)
```

```python
mid_salary = [ float(((v[::-1])[2:])[::-1]) for v in mid_salary ]
print(mid_salary)
```

```python
print(len(mid_salary))
```

```python
middle_salary = pd.DataFrame()
middle_salary["year"] = [1996+i for i in range(25)]
middle_salary["mid salary, hryvnias"] = mid_salary
middle_salary.to_csv("middle_salary_primary.csv")
```

Fertility rate

```python
fertility_rate_data = get_data("https://uk.wikipedia.org/wiki/Народжуваність_в_Україні").find_all("table")
fertility_rate_data = fertility_rate_data[7].find_all("td")
fertility_rate_data[14*2:]
```

```python
fertility_rate_data = [value.text for value in fertility_rate_data[14*2:] if fertility_rate_data[14*2:].index(value)%2 ]
fertility_rate_data
```

```python
# element = fertility_rate_data[-1].replace('\n', "").replace(",", ".").replace("[14]", "")
element = fertility_rate_data[-1].replace(",", ".")[:4]
fertility_rate_data = [float(value[2:].replace('\n', "").replace(",", ".").replace("[14]", "")) for value in fertility_rate_data[:-1]]

fertility_rate_data = fertility_rate_data + [float(element)]
fertility_rate_data
```

```python
fertility_rate_data_save = pd.DataFrame()
fertility_rate_data_save["year"] = [1996+i for i in range(26)]
fertility_rate_data_save["fertility rate"] = fertility_rate_data
fertility_rate_data_save
```

```python
fertility_rate_data_save.to_csv("fertility_rate_data_primary.csv")
```

Mortality rate

```python
mortality_rate_data = get_data("https://uk.wikipedia.org/wiki/Смертність_в_Україні").find_all("table")
mortality_rate_data = mortality_rate_data[3]
mortality_rate_data
```

```python
tds = mortality_rate_data.find_all("td")
tds = [ (int(tds[5*i].text[:4]), float(tds[5*i+2].text.replace(",", "."))) for i in range(30+1)] #range(int(len(tds))) ]
tds
```

```python
mortality_rate_data_save = pd.DataFrame(tds)
mortality_rate_data_save.columns = ["year", "mortality rate"]
mortality_rate_data_save
```

```python
mortality_rate_data_save.to_csv("mortality_rate_primary.csv")
```

GDP

```python
GDP_data = get_data("https://uk.wikipedia.org/wiki/Валовий_внутрішній_продукт_України").find_all("table")
GDP_data = GDP_data[3].find_all("td")
GDP_data = GDP_data[4*8:]
GDP_data = GDP_data[5*8:]
GDP_data
```

```python
GDP_data_list = [ (int(GDP_data[8*i].text.replace("*", "")), float(GDP_data[8*i+5].text[:-2].replace(",", "."))) for i in range(25+1) ]
GDP_data_list
```

```python
GDP_data_save = pd.DataFrame(GDP_data_list)
GDP_data_save.columns = ["year", "GDP"]
GDP_data_save
```

```python
GDP_data_save.to_csv("GDP_primary.csv")
```
