```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

# Part 1
**Data reading and data package forming**

```python
data1 = pd.read_csv("data/Consumer_price_indices_for_food_and_services.csv", sep=";")
data2 = pd.read_csv("data/population.csv", sep=";")
```

```python
data1
```

```python
data2
```

```python
data2 = data2.take(data2.index[11:])
data2.index = data1.index.copy()
```

```python
data = pd.DataFrame()

for col in data1.columns[1:]:
    v = data1[col].copy().to_list()
    v = [ float(el.replace(",", ".")) for el in v ]
    
    data[col] = v
for col in data2.columns[1:]:
    v = data2[col].copy().to_list()
    v = [ float(el.replace(",", ".")) for el in v ]
    
    data[col] = v
data.index = data1["Year"]
data = data.astype("float")

del data1
del data2

data = data.astype("float")
```

```python
data
```

```python
data.info()
```

```python
presents = data["Total_present_population_thnds"]
data = data.drop("Total_present_population_thnds", axis="columns")

residents = data["Total_resident_population_thnds"]
data = data.drop("Total_resident_population_thnds", axis="columns")
```

# Part 2
**EDA**

Firstly, let's calculate some statistics.

```python
# t-test for relation
# calculation p-values for presents number
# Significance level: 0.01

for col in data.columns:
    print(f"For {col} p-value={stats.ttest_rel(data[col], presents, alternative='two-sided')[1]}")
```

```python
# t-test for relation
# calculation p-values for residents number
# Significance level: 0.01

for col in data.columns:
    print(f"For {col} p-value={stats.ttest_rel(data[col], residents, alternative='two-sided')[1]}")
```

**We see, that hypothesis, that selected parameters are insignificant, is an unlikely for both parameters we will try to predict.** So, let's calculate Pearson coefficients.

```python
# calculation r coefficients for presents number

for col in data.columns:
    print(f"For {col} r={data[col].corr(presents)}")
```

```python
# calculation r coefficients for residents number

for col in data.columns:
    print(f"For {col} r={data[col].corr(residents)}")
```

We see, that residents number and presents number are bad-collerated with selected parameters. This indicates, that for prediction we may try to use linear regression. But! There's additional reason to check, **are selected predictors well-correlated with each other or no?**

```python
# to check it, let's plot correlation matrix
sns.heatmap(data.corr(), vmin=-1, vmax=1)
```

```python
# let's sellect well-correlated predictors

cols = list(data.columns)
for i in range(len(cols)-1):
    for j in range(i+1, len(cols)):
        if abs(data[cols[i]].corr(data[cols[j]]))>=0.9:
            print(f"{cols[i]} and {cols[j]}")
```

We see, that some predictors are well-correlated. So, it would be good to drop it.

```python
cols_to_drop = [
    "Consumer_price_indices",
    "Furnishings_household_equipment_and_routine_maintenance_of_the_house",
    "Miscellaneous_goods_and_services",
    "Clothing_and_footwear"
]

data = data.drop(cols_to_drop, axis="columns")
cols = list(data.columns)
for i in range(len(cols)-1):
    for j in range(i+1, len(cols)):
        if abs(data[cols[i]].corr(data[cols[j]]))>=0.9:
            print(f"{cols[i]} and {cols[j]}")
```

We dropped well-correlated predictors. So, **let's make more detailed analysis of behavior of dependencies**.

```python
# X = data.take(data.index[:-1]).copy()
index = data.index
X = data.T[index[:-1]].T.copy()
```

**Firstly, let's experiment with ElasticNet.**

```python
from sklearn.linear_model import ElasticNet
from sklearn.experimental import enable_halving_search_cv
```

```python
elastic = ElasticNet()

grid = {
    "alpha" : [i/10 for i in range(1, 11)],
    "l1_ratio" : [i/10 for i in range(1, 11)]
}
selection = model_selection.GridSearchCV(elastic, grid)
selection.fit(X, presents[1:])

print(selection.get_params())
```

We see, that optimal values of needed parameters of ElasticNet in sklearn is given by default.

```python
predictors_prediction_models = {
    col : ElasticNet().fit(data.drop(col, axis=1).T[index[:-1]].T, data[col][1:]) for col in data.columns
}
```

To use prediction models like neural networks, we may not have enough data. Using models like decision trees or polynomial regression may give not more realistic prediction. But, **what if we try to build more specific prediction model?**

In some literature, we may find prediction method, based on usage of linear filters
$$ x_m=\sum_{k=1}^{n} a_{k}x_{m-k} $$
or something similar.
What if we try to build a prediction model like this?

```python
len(presents)
```

```python
len(residents)
```

```python
new_presents = presents.copy()
new_residents = residents.copy()
```

```python
type(new_presents)
```

```python
# let's n:=11

# ak coefficients vectors
presents_a = [] 
residents_a = [] 
# matrix of values
presents_A = [] 
residents_A = [] 
# b vectors
presents_b = []
residents_b = []

# calculation of coefficients

presents_b = new_presents.to_list()[11:]
residents_b = new_residents.to_list()[11:]

presents_A = np.array([ new_presents.to_list()[r:r+11] for r in range(11) ])
residents_A = np.array([ new_residents.to_list()[r:r+11] for r in range(11) ])

presents_a = np.linalg.solve(presents_A, presents_b)
residents_a = np.linalg.solve(residents_A, residents_b)
```

```python
presents_a
```

```python
residents_a
```

```python
# let's build and visualize a prediction

new_presents = presents.to_list()
new_residents = residents.to_list()

# prediction
for year in range(2023, 2030+1):
    new_presents.append( sum(np.array(new_presents[-11:])*presents_a) )
    new_residents.append( sum(np.array(new_residents[-11:])*residents_a) )

# visualization
plt.plot(range(2001, 2022+1), new_presents[:-8], "b", label="Presents' number")
plt.plot(list(range(2022, 2030+1)), new_presents[-9:], "--b", label="Presents' number prediction using linear filter model")

plt.plot(range(2001, 2022+1), new_residents[:-8], "r", label="Residents' number")
plt.plot(range(2022, 2030+1), new_residents[-9:], "--r", label="Residents' number prediction using linear filter model")

plt.grid()
plt.legend()
plt.show()
```

Prediction is unrealistic. But, what if we try to predict not values, but logarithms from it?

```python
# ak coefficients vectors
presents_a = [] 
residents_a = [] 
# matrix of values
presents_A = [] 
residents_A = [] 
# b vectors
presents_b = []
residents_b = []

# calculation of coefficients
new_presents = presents.copy()
new_residents = residents.copy()

presents_b = new_presents.to_list()[11:]
presents_b = np.log(presents_b)
residents_b = new_residents.to_list()[11:]
residents_b = np.log(residents_b)

presents_A = np.array([ new_presents.to_list()[r:r+11] for r in range(11) ])
residents_A = np.array([ new_residents.to_list()[r:r+11] for r in range(11) ])

presents_a = np.linalg.solve(presents_A, presents_b)
residents_a = np.linalg.solve(residents_A, residents_b)

new_presents = presents.to_list()
new_residents = residents.to_list()

# prediction
for year in range(2023, 2030+1):
    pred_presents = np.exp( sum(np.array(new_presents[-11:])*presents_a) )
    pred_residents = np.exp( sum(np.array(new_residents[-11:])*residents_a) )
    
    new_presents.append(pred_presents)
    new_residents.append( sum(np.array(new_residents[-11:])*residents_a) )

# visualization
plt.plot(range(2001, 2022+1), new_presents[:-8], "b", label="Presents' number")
plt.plot(list(range(2022, 2030+1)), new_presents[-9:], "--b", label="Presents' number prediction using linear filter model")

plt.plot(range(2001, 2022+1), new_residents[:-8], "r", label="Residents' number")
plt.plot(range(2022, 2030+1), new_residents[-9:], "--r", label="Residents' number prediction using linear filter model")

plt.grid()
plt.legend()
plt.show()
```

We see, that **linear filer as prediction model gives unrealistic prediction here**. But, the idea of model is not bad. So, **what if we use polynomial extrapolation?**

```python
# let's degree of polynomials n:=20
#
# We will consider, that every parameter in formed dataset depends of number of seen year
# To realize extrapolation for population we will build extrapolation polynolials as interpolation polynomials for every else parameters and
# will use weigth sum of it

# ak coefficients vectors
presents_a = np.zeros(21) 
residents_a = np.zeros(21) 

# calculation of coefficients vectors
a_matrix = []

for col in data.columns:
    v = data[col].to_numpy()
    A = np.array([ [pow(t, k) for k in range(21)] for t in range(1, 22) ])
    a_matrix.append( np.linalg.solve(A.astype("float"), v[1:]) )
    # print(A.astype("float"))

a_matrix = np.array(a_matrix)
a_matrix.shape

for i in range(9):
    presents_a += presents_prediction_model.coef_[i]*a_matrix[i]
    residents_a += residents_prediction_model.coef_[i]*a_matrix[i]

# prediction
new_presents = presents.to_list()
new_residents = residents.to_list()

# prediction
for year in range(2023, 2030+1):
    new_presents.append( sum(np.array([ pow(year-2000, k) for k in range(21) ])*presents_a) )
    new_residents.append( sum(np.array([ pow(year-2000, k) for k in range(21) ])*residents_a) )

# visualization
plt.plot(range(2001, 2022+1), new_presents[:-8], "b", label="Presents' number")
plt.plot(list(range(2022, 2030+1)), new_presents[-9:], "--b", label="Presents' number prediction using extrapolation")

plt.plot(range(2001, 2022+1), new_residents[:-8], "r", label="Residents' number")
plt.plot(range(2022, 2030+1), new_residents[-9:], "--r", label="Residents' number prediction using extrapolation")

plt.grid()
plt.legend()
plt.show()
```

We see, that extrapolation as prediction method gives more unrealistic and incorrect prediction, than linear filters.
