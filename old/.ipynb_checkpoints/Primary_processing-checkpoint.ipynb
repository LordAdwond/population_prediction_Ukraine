```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
middle_salary_p = pd.read_csv("middle_salary_primary.csv")[["year", "mid salary, hryvnias"]]
population_info_p = pd.read_csv("population_info_primary.csv")[["year", "population (thousands)"]]
fertility_rate_p = pd.read_csv("fertility_rate_data_primary.csv")[["year", "fertility rate"]]
mortality_rate_p = pd.read_csv("mortality_rate_primary.csv")[["year", "mortality rate"]]
GDP_p = pd.read_csv("GDP_primary.csv")[["year", "GDP"]]
```

Data Processing

```python
len(middle_salary_p.index)
```

```python
len([(middle_salary_p.loc[i]["year"]>=1996 and middle_salary_p.loc[i]["year"]<=2020) for i in middle_salary_p.index.to_list()])
```

```python
index = middle_salary_p.index
index = [(middle_salary_p.loc[i]["year"]>=1996 and middle_salary_p.loc[i]["year"]<=2020) for i in index.to_list()]
middle_salary = middle_salary_p[index]
middle_salary = middle_salary.dropna()
middle_salary.to_csv("middle_salary.csv")

index = population_info_p.index
index = [(population_info_p.loc[i]["year"]>=1996 and population_info_p.loc[i]["year"]<=2020) for i in index]
population_info = population_info_p[index]
population_info = population_info.dropna()
population_info.to_csv("population_info.csv")

index = fertility_rate_p.index
index = [(fertility_rate_p.loc[i]["year"]>=1996 and fertility_rate_p.loc[i]["year"]<=2020) for i in index]
fertility_rate = fertility_rate_p[index]
fertility_rate = fertility_rate.dropna()
fertility_rate.to_csv("fertility_rate.csv")

index = mortality_rate_p.index
index = [(mortality_rate_p.loc[i]["year"]>=1996 and mortality_rate_p.loc[i]["year"]<=2020) for i in index]
mortality_rate = mortality_rate_p[index]
mortality_rate = mortality_rate.dropna()
mortality_rate.to_csv("mortality_rate.csv")

index = GDP_p.index
index = [(GDP_p.loc[i]["year"]>=1996 and GDP_p.loc[i]["year"]<=2020) for i in index]
GDP = GDP_p[index]
GDP = GDP.dropna()
GDP.to_csv("GDP.csv")
```
