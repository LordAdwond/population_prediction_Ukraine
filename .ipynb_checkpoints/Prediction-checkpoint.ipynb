Let try to build prediction model of R^5k -> R^5 type.

Linear reccurent model

```python
Data = data[data.columns[1:]]
Data
```

```python
means = Data.mean()
stds = Data.std()

Data = (Data-means)/stds
Data
```

```python
# squared deviation for k-th parameter
def E(a : list, k : int):
    global Data
    a = np.array(a)
    return sum([ pow( Data[Data.columns[k]][n+2] - sum(a[:5]*Data.iloc[n+1]) - sum(a[5:]*Data.iloc[n+1]), 2 ) for n in range(len(Data)-2) ])

# derivative by i-th coefficient for k-th parameter's squared deviation
def dE_dai(a : list, i : int, k : int, h=0.01):
    new_a = a.copy()
    new_a[i] += h
    
    return (E(new_a, k) - E(a, k))/h

# mimimizing of squared deviation for k-th parameter
def error_minimizing_k(a : list, k : int, eps=0.001):
    A = np.array(a)
    dA = np.array([ dE_dai(a, i, k) for i in range(10) ])
    A = A - dA
    
    if( np.linalg.norm(dA)>=eps ):
        return error_minimizing_k(A.tolist(), k)
    else:
        return A.tolist()
```

```python
A = [[1 for i in range(10)] for j in range(5)]
for i in range(5):
    A[i] = error_minimizing_k(A[i], i)
A
```

```python
def salary_next(x : list, a=A[0]):
    return sum( np.array(x)*np.array(a) )
def population_next(x : list, a=A[1]):
    return sum( np.array(x)*np.array(a) )
def fertility_next(x : list, a=A[2]):
    return sum( np.array(x)*np.array(a) )
def mortality_next(x : list, a=A[3]):
    return sum( np.array(x)*np.array(a) )
def GDP_next(x : list, a=A[4]):
    return sum( np.array(x)*np.array(a) )
```

```python
x0 = Data.iloc[0].tolist()
x1 = Data.iloc[1].tolist()
prediction = [x0[1], x1[1]]
for i in range(len(Data)):
    x2  = [salary_next(x0+x1), population_next(x0+x1), fertility_next(x0+x1), mortality_next(x0+x1), GDP_next(x0+x1)]
    x0 = x1.copy()
    x1 = x2.copy()
    
    prediction.append(x2[1])

prediction = np.array(prediction)
plt.plot(Data["population (thousands)"], label="real data")
plt.plot(prediction*means[1] + stds[1], label="prediction")
plt.grid()
plt.legend()
plt.show()
```

As we see, models like yk(n+2)=a1k\*y1(n+1)+...+a5k\*y5(n+1)+b1k\*y1(n)+...+b5k\*y5(n) for k from 1 to 5 may give very incorrect results. So we need some other prediction model of R^5k -> R^5 type.
